## 1. Problem Definition

In this case, the problem we aim to solve is: **Given one year of transaction data, can we segment customers into meaningful groups according to their spending habits?**

There are numerous approaches to segmenting business customers, each of which should be carefully tailored to meet specific business needs. For the purpose of simplicity, this project employs a well-known method known as RFM, which stands for Recency, Frequency, and Monetary value. This method categorizes customers based on how recently they have made transactions, the frequency of their purchases, and the total amount they have spent.

Recency: Measures how recently a customer made a purchase. Customers who have purchased recently are often more engaged and potentially more receptive to new offers and products. This metric helps in identifying those who are more likely to respond to immediate marketing activities.

Frequency: Measures how often a customer makes a purchase within a specific time frame. High-frequency customers are typically more loyal and potentially more valuable in the long term. Understanding frequency helps in recognizing and rewarding loyalty or encouraging increased engagement from less frequent shoppers.

Monetary Value: Measures how much money a customer spends over a given period. This metric helps in identifying high-value customers who contribute significantly to the revenue and might deserve special attention or perks to maintain or enhance their loyalty.

## 3. Features engineering

### Grouped by customer

In order to segment the customers, new features need to be created and analyzed to group the customers by their characteristics based on RFM method.

#### Monetary value: Calculating spend per customer

```python
# Calculate the total spend for each transaction first
customers_orders['Transaction_Total'] = (customers_orders['Quantity'] * customers_orders['Avg_Price']) + customers_orders['Delivery_Charges'] + customers_orders['GST']

# Now, calculate the total spend per customer
total_spend_per_customer = customers_orders.groupby('CustomerID')['Transaction_Total'].sum().reset_index()

#Display the new feature
total_spend_per_customer.head()
```

#### Monetary value: Calculating the average transaction value per customer
This feature is intended solely for exploratory analysis and will not be incorporated into the model's development.

```python
# Calculate the average value per customer
average_transaction_value_per_customer = customers_orders.groupby("CustomerID")["Transaction_Total"].mean().reset_index()

# Rename the new column
average_transaction_value_per_customer.rename(columns={'Transaction_Total': 'Average_Transaction_Value'}, inplace=True)

# Display the new column head
average_transaction_value_per_customer.head()
```

#### Frequency: Calculating number of transactions per customer

```python
# Next, calculate the number of transactions per customer
transactions_per_customer = customers_orders.groupby('CustomerID')['Transaction_ID'].nunique().reset_index()

# Rename the new column
transactions_per_customer = transactions_per_customer.rename(columns={'Transaction_ID': 'Transaction_Count'})

# Display the new column head
transactions_per_customer.head()
```

#### Recency: Calculating the most recent transaction of each customer

```python
# Ensure the 'Transaction_Date' is in datetime format
customers_orders['Transaction_Date'] = pd.to_datetime(customers_orders['Transaction_Date'])

# Define the reference date as 1 January 2020
reference_date = pd.Timestamp('2020-01-01')

# Calculate days since last purchase for each transaction with respect to the reference date
customers_orders['Days_Since_Last_Purchase'] = (reference_date - customers_orders['Transaction_Date']).dt.days

# Now calculate the most recent purchase per customer (i.e., minimum days since last purchase)
days_since_last_purchase_per_customer = customers_orders.groupby('CustomerID')['Days_Since_Last_Purchase'].min().reset_index()

# Rename the column for clarity
days_since_last_purchase_per_customer.rename(columns={'Days_Since_Last_Purchase': 'Days_Until_2020'}, inplace=True)

# Display the new column head
days_since_last_purchase_per_customer.head()
```

### Choosing and applying transformation
There are several transformations to choose, so some of them will be compared to find the best method.

```python
def visualize_transformations_plotly(results, feature_name, title_font_size=14):
    # Determine the number of transformation results to plot
    num_transformations = len(results)
    
    # Calculate the number of columns for the subplots, considering up to 3 transformations per row for readability
    cols = min(num_transformations, 3)
    rows = (num_transformations + cols - 1) // cols  # Calculate rows needed, ensuring we cover all transformations
    
    # Create a subplot figure
    fig = make_subplots(rows=rows, cols=cols, subplot_titles=[f"{name}\nSkew: {data['Skewness']:.2f}" for name, data in results.items()],
                        horizontal_spacing=0.05, vertical_spacing=0.1)
    
    # Index for tracking the current plot position
    current_plot = 1
    for name, data in results.items():
        # Determine the current row and column
        row = (current_plot - 1) // cols + 1
        col = (current_plot - 1) % cols + 1
        
        transformed_data_values = data['Transformed Feature'].values
        fig.add_trace(
            go.Histogram(x=transformed_data_values, nbinsx=30, name=name),
            row=row, col=col
        )
        current_plot += 1
    
    # Update layout to adjust for the figure title and make the layout tighter
    fig.update_layout(title_text=f'Transformations of {feature_name}', title_x=0.5, showlegend=False, height=400*rows)
    fig.update_traces(opacity=0.75)  # Update trace options if needed
    
    # Update subplot titles font size
    fig.update_annotations(font=dict(size=title_font_size))
    
    # Show the figure
    fig.show()
```

```python
def apply_transformations(feature):
    # Ensure the input feature is a pandas Series
    feature = pd.Series(feature)
    
    transformations = {'Original': feature,
                       'Log': np.log1p(feature),
                       'Square root': np.sqrt(feature),
                       'Box-Cox': boxcox1p(feature, boxcox_normmax(feature.dropna() + 1)),
                       'Yeo-Johnson': pd.Series(PowerTransformer(method='yeo-johnson').fit_transform(feature.dropna().values.reshape(-1,1)).flatten()),
                       'Inverse': 1 / (feature + 0.0001),  # Adding a small constant to avoid division by zero
                       'Cube root': np.cbrt(feature)
                      }
    
    results = {}
    for name, transformed_feature in transformations.items():
        # Convert NumPy arrays to pandas Series for compatibility with dropna
        if isinstance(transformed_feature, np.ndarray):
            transformed_feature = pd.Series(transformed_feature)
        
        skewness = skew(transformed_feature.dropna())  # Now works as transformed_feature is a pandas Series
        results[name] = {'Skewness': skewness, 'Transformed Feature': transformed_feature}
    
    return results
```

```python
transaction_transformed_result = apply_transformations(features_by_customer["Transaction_Total"])
visualize_transformations_plotly(transaction_transformed_result, "Transaction_Total")
```

```python
avg_transaction_transformed_result = apply_transformations(features_by_customer["Average_Transaction_Value"])
visualize_transformations_plotly(avg_transaction_transformed_result, "Average_Transaction_Value")
```

```python
avg_transaction_transformed_result = apply_transformations(features_by_customer["Transaction_Count"])
visualize_transformations_plotly(avg_transaction_transformed_result, "Transaction_Count")
```

Observing from the skewness, the chosen transformations for each feature are as following:

- Transaction total --> Yeo-Johnson transformation
- Average transaction value --> Box-Cox transformation
- Transaction count --> Box-Cox transformation

```python
# Make a reusable function for applying chosen transformation

def apply_selected_transformation(df, features, transformation_type):
    """
    Applies a selected transformation to specified features in the DataFrame.
    
    Parameters:
    - df: pandas DataFrame containing the data.
    - features: list of strings, names of the features to transform.
    - transformation_type: string, type of transformation to apply. 
      Options include 'log', 'square_root', 'yeo-johnson'.
      
    Returns:
    - Transformed DataFrame.

    Example:
    df = pd.DataFrame({
    'Feature1': [0, 1, 2, 3, 4, 5],
    'Feature2': [10, 20, 30, 40, 50, 60],
    'Feature3': [100, 200, 300, 400, 500, 600]
    })
    
    features_to_transform = ['Feature1', 'Feature2']
    transformation_type = 'log'  # Choose 'log', 'square_root', or 'yeo-johnson'
    
    transformed_features = apply_selected_transformation(df, features_to_transform, transformation_type)
    print(transformed_features)
    """
    # Copy the DataFrame to avoid modifying the original data
    transformed_features = df.copy()
    
    for feature in features:
        # Ensure the feature is a pandas Series
        data = transformed_features[feature]
        
        if transformation_type == 'log':
            transformed_features[feature] = np.log1p(data)
        elif transformation_type == 'square_root':
            transformed_features[feature] = np.sqrt(data)
        elif transformation_type == 'box-cox':
            # Box-Cox requires positive data
            transformed_features[feature] = boxcox1p(data, boxcox_normmax(data.dropna() + 1))
        elif transformation_type == 'yeo-johnson':
            # Yeo-Johnson works with both positive and negative data
            pt = PowerTransformer(method='yeo-johnson')
            transformed_feature = pt.fit_transform(data.values.reshape(-1, 1))
            transformed_features[feature] = transformed_feature.flatten()
        elif transformation_type == 'inverse':
            transformed_features[feature] = 1 / (data + 0.0001)  # Adding a small constant to avoid division by zero
        elif transformation_type == 'cube_root':
            transformed_features[feature] = np.cbrt(data)
        else:
            raise ValueError("Unsupported transformation type provided.")
    
    return transformed_features
```

```python
# Apply transformation to selected features
transformed_features = apply_selected_transformation(features_by_customer, ["Transaction_Total", "Transaction_Count"], "yeo-johnson")
transformed_features = apply_selected_transformation(features_by_customer, ["Average_Transaction_Value"], "box-cox")

transformed_features_1 = transformed_features[["Transaction_Total", "Transaction_Count", "Days_Until_2020"]]
```

```python
transformed_features_1.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Assuming transformed_features is your DataFrame after transformations
features = transformed_features_1.columns  # Or a subset of columns if not all are relevant
X = transformed_features_1[features]

scaler = StandardScaler().fit(X)
X_scaled = scaler.transform(X)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(init='k-means++',n_clusters=i,max_iter=500,random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
optimal_clusters = 4  # This is an example; use your own method to find the optimal number
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans.fit(X_scaled)
cluster_labels = kmeans.labels_
```

```python
transformed_features_1['Cluster'] = cluster_labels

# Explore the cluster's characteristics
for i in range(optimal_clusters):
    print(f"Cluster {i}:")
    cluster_data = transformed_features_1[transformed_features_1['Cluster'] == i]
    # Perform your analysis, e.g., describe the cluster, mean of features, etc.
    print(cluster_data.describe())
    # You might want to visualize feature distributions within the cluster, etc.
```
